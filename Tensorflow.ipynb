````python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.arange(-5,5,.01)
````

````python
def plot(func,yaxis = (-1.4,1.4)):
  plt.ylim(yaxis)
  plt.locator_params(nbins=5)
  plt.xticks(fontsize = 14)
  plt.yticks(fontsize= 14)
  plt.axhline(lw=1,c='black') #horizontal line across the axis
  plt.axvline(lw=1,c='black') #horizontal line across the axis 
  plt.grid(alpha=0.4,ls='-.') #alpha learning rate
  plt.box(on=None)
  plt.plot(x,func(x),c='r',lw=3)
````

# **Rectified linear Units/relu**
0 for x<=0
1 for x>=0
```

```

````python
relu = np.vectorize(lambda x: x if x>=0 else 0,otypes=[np.float])
plot(relu)
````

# **Leaky relu**


````python
leaky_relu = np.vectorize(lambda x: max(0.1*x,x),otypes=[np.float])
plot(leaky_relu)
````

````python
x = tf.Variable(1.0)
def f(x):
  y = x**+2*x**2-5
  return y
````

````python
print(f(x))
````

````python
with tf.GradientTape() as tape:
  y = f(x)
g_x = tape.gradient(y,x) #np.gradient = dy/dx
print(g_x)
````

````python
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams['figure.figsize'] = [9,6]
````

````python
x= tf.linspace  (-2,2,201)
x= tf.cast(x, tf.float32)
````

````python
def f(x):
  y = x**2 +2*x - 5
  return y

y = f(x) + tf.random.normal(shape = [201])
plt.plot(x.numpy(), y.numpy(), '.', label = 'Data')
plt.plot(x,f(x),label = 'Ground Truth')
plt.legend();
````
