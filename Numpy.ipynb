# **Random Function**

```python
random.rand(100,100) # genrates a random array from 0 to 1
```

```python
print(random.rand(5),"\n\n\n")
print(random.rand(5,6),"\n\n\n")
#random.rand((2,3)) #throws an error if tupple passed
print(random.random((2,3))) # can pass tupple also
```

```python
print(random.randint(low=55,high=999,size=(3,5,),dtype=int32)) # low--> start ,hig-->end,size-->tupple/int for size ,dtype-->(optional)
```

```python
print(random.choice([6,7,8,9,100],(2,3))) 
"""parameters 
a--> int or list 
 size = int or tupple 
replace = T/F--> if you want to use element from passed a manny times  
p --> probabilties of each instances passed in the a"""
```

# **Slicing in numpy**


```python
random.random((3,3))
```

```python
b = array([[2,3,5,4],[5,3,5,1],[1,3,5,7]],dtype= int32)
print(b[::2,::3])
print(b[::2,[0,-1]])
```
