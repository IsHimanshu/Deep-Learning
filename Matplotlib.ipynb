```python
from matplotlib import pyplot as plt
import numpy as np
#x = np.linspace(2*pi)
```

```python
x = range(0,10,3)
y = [3+i + 2 for i in x]
plt.plot(x,y,marker='d')
plt.xlabel('Ichi')
plt.ylabel('Ni')
plt.title('Random tittle')
```

```python
#formatting 
plt.plot(x,y,'ro')
```

```python
#formating diff points
y2 = [i**2+10 for i in x]
plt.plot(x,y,'r--',y2,'bs')
```

```python
#scater plot 
plt.scatter(x,y2)
```

```python
#ploting multiple graphs in asingle
group = ['Naruto','One Piece','Bleach']
plt.figure(figsize=(10,3))
values = [5000,100000,1000]
plt.subplot(121)
plt.bar(group,values)
plt.subplot(122)
plt.scatter(group,values)
```
